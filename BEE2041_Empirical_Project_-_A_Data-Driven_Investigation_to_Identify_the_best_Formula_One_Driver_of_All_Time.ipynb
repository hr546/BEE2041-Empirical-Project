```python
### BEE2041: Data Science in Economics - Empirical Project
### A Data-Driven Investigation to Identify the Best Formula One Driver of All Time
### Written by Hamish Roberts (730102814)

# The question of who the greatest Formula One driver of all time is has long been passionately debated, with Michael
# Schumacher, Lewis Hamilton, Aryton Senna, and now Max Verstappen all boasting records that put them in contention. Whilst
# different fans within the sport will all attempt to argue that their favourite driver is better than the others, this is 
# a question that will never be able to be definitively answered due to the fact that each of these drivers were in the sport
# in different eras, and never raced against each other in equal machinery. The presence of evolving regulations and points 
# scoring systems, as well as the ever-increasing number of races on the calender and constantly improving technical prowess
# of the cars, makes comparisons between the drivers even more difficult. Whilst it is impossible to go back in time to race
# these drivers against each other in a controlled experiment, this investigation aims to cut through the litany of 
# extraneous variables to conclude who the best Formula One driver of all time is by taking a data-driven analytical approach.
```

```python
# Bar chart of wins
# https://pitwall.app/records/driver-wins?page=1

# Creating a bar chart to demonstrate the number of wins for each driver
# For conciseness, only the drivers within the top 50 for number of wins are included

from urllib import request, error
from bs4 import BeautifulSoup
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Webscraping the website containing the number of wins for each driver (1-25)
url = 'https://pitwall.app/records/driver-wins?page=1'

response = BeautifulSoup(request.urlopen(url), 'html.parser')

TAB = response.find('table')

Wins_df = pd.DataFrame(columns = ['Position','Driver Name','First Win','Last Win','Driver Races','Driver Wins'])

rows = TAB.find_all('tr')

# Inserting the webscraped data into the dataframe
for row in rows[1:]:
    cells = row.find_all('td')
    row_contents = []
    for cell in cells:
        row_contents.append(cell.text.strip())
    length = len(Wins_df)
    Wins_df.loc[length] = row_contents

# Webscraping the website containing the number of wins for the next 25 drivers
url = 'https://pitwall.app/records/driver-wins?page=2'

response = BeautifulSoup(request.urlopen(url), 'html.parser')

TAB = response.find('table')

rows = TAB.find_all('tr')

# Inserting this new data into the same dataframe
for row in rows[1:]:
    cells = row.find_all('td')
    row_contents = []
    for cell in cells:
        row_contents.append(cell.text.strip())
    length = len(Wins_df)
    Wins_df.loc[length] = row_contents

# Dropping unnecessary columns from the dataframe
Wins_df = Wins_df.drop('First Win', axis=1)
Wins_df = Wins_df.drop('Last Win', axis=1)
Wins_df = Wins_df.drop('Position', axis=1)

# Changing the 'Driver Wins' column from a string to an integer
Wins_df['Driver Wins'] = Wins_df['Driver Wins'].astype(int)

# Plotting the bar chart for number of wins
Wins_df.plot.barh(x = 'Driver Name', y = 'Driver Wins', fontsize = 6)
plt.gca().invert_yaxis()
```

```python
# Bar chart of points
# https://pitwall.app/records/points-scored

# Creating a bar chart to demonstrate the number of points for each driver
# Only the drivers within the top 50 for points have been included

# Webscraping the website containing the number of points for each driver (1-50)
url = ' https://pitwall.app/records/points-scored'

response = BeautifulSoup(request.urlopen(url), 'html.parser')

TAB = response.find('table')

Points_df = pd.DataFrame(columns = ['Position','Driver Name','First Points','Last Points','Driver Races','Driver Points'])

rows = TAB.find_all('tr')

# Inserting the webscraped data into the dataframe
for row in rows[1:]:
    cells = row.find_all('td')
    row_contents = []
    for cell in cells:
        row_contents.append(cell.text.strip())
    length = len(Points_df)
    Points_df.loc[length] = row_contents

# Dropping unnecessary columns from the dataframe
Points_df = Points_df.drop('First Points', axis=1)
Points_df = Points_df.drop('Last Points', axis=1)
Points_df = Points_df.drop('Position', axis=1)

# Changing the 'Driver Points' column from a string to an integer by first removing the commas
Points_df = Points_df.replace(",","", regex=True)
Points_df['Driver Points'] = Points_df['Driver Points'].astype(float)

# Plotting the bar chart for number of points
Points_df.plot.barh(x = 'Driver Name', y = 'Driver Points', fontsize = 6)
plt.gca().invert_yaxis()
```

```python
# Bar chart of world championships
# https://pitwall.app/records/driver-championships

# Creating a bar chart to demonstrate the number of world championships won by each driver
# Only 34 drivers are included, as only 34 drivers have ever won a drivers' world championship

# Webscraping the website containing the number of points for each driver (1-34)
url = ' https://pitwall.app/records/driver-championships'

response = BeautifulSoup(request.urlopen(url), 'html.parser')

TAB = response.find('table')

Championships_df = pd.DataFrame(columns = ['Position','Driver Name','Races','Wins','Championships'])

rows = TAB.find_all('tr')

# Inserting the webscraped data into the dataframe
for row in rows[1:]:
    cells = row.find_all('td')
    row_contents = []
    for cell in cells:
        row_contents.append(cell.text.strip())
    length = len(Championships_df)
    Championships_df.loc[length] = row_contents

# Dropping unnecessary columns from the dataframe
Championships_df = Championships_df.drop('Wins', axis=1)
Championships_df = Championships_df.drop('Races', axis=1)
Championships_df = Championships_df.drop('Position', axis=1)

# Changing the 'Driver Pole Positions' column from a string to an integer
Championships_df['Championships'] = Championships_df['Championships'].astype(int)

# Plotting the bar chart for number of championships
Championships_df.plot.barh(x = 'Driver Name', y = 'Championships', fontsize = 8)
plt.gca().invert_yaxis()
```

```python
# Histogram of percentage of race wins
# https://pitwall.app/records/driver-wins?page=1

# Converting "Driver Races" to an integer
Wins_df['Driver Races'] = Wins_df['Driver Races'].astype(int)

# Adding additional column to dataframe containing the percentage of races won by each driver
Wins_df["Win Percentage"] = (Wins_df["Driver Wins"]/Wins_df["Driver Races"]) * 100

# Sorting dataframe so the highest percentage is listed first
Wins_df = Wins_df.sort_values(by=["Win Percentage"])

# Plotting the bar chart for percentage of race wins
Wins_df.plot.barh(x = 'Driver Name', y = 'Win Percentage', fontsize = 6.5)

# Displaying dataframe - limited to top 10
display(Wins_df.tail(10))
```

```python
# Histogram of percentage of podiums
# https://pitwall.app/records/driver-podiums

# Converting "Driver Races" to an integer
Podiums_df['Driver Races'] = Podiums_df['Driver Races'].astype(int)

# Adding additional column to dataframe containing the percentage of podiums achieved by each driver
Podiums_df["Podium Percentage"] = (Podiums_df["Driver Podiums"]/Podiums_df["Driver Races"]) * 100

# Sorting dataframe so the highest percentage is listed first
Podiums_df = Podiums_df.sort_values(by=["Podium Percentage"])

# Plotting the bar chart for percentage of race wins
Podiums_df.plot.barh(x = 'Driver Name', y = 'Podium Percentage', fontsize = 6.5)

# Displaying dataframe - limited to top 10
display(Podiums_df.tail(10))
```

```python
# Histogram of percentage of pole positions
# https://pitwall.app/records/driver-pole-positions

# Converting "Driver Races" to an integer
Poles_df['Driver Races'] = Poles_df['Driver Races'].astype(int)

# Adding additional column to dataframe containing the percentage of podiums achieved by each driver
Poles_df["Pole Position Percentage"] = (Poles_df["Driver Pole Positions"]/Poles_df["Driver Races"]) * 100

# Sorting dataframe so the highest percentage is listed first
Poles_df = Poles_df.sort_values(by=["Pole Position Percentage"])

# Plotting the bar chart for percentage of race wins
Poles_df.plot.barh(x = 'Driver Name', y = 'Pole Position Percentage', fontsize = 6.5)

# Displaying dataframe - limited to top 10
display(Poles_df.tail(10))
```
