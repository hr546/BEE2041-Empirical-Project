### Section 1: Analysing Driver Statistics of Wins, Podiums, Pole Positions, Points and Championships

```python
# Creating a bar chart to demonstrate the number of wins for each driver
# For conciseness, only the drivers within the top 50 for number of wins are included

from urllib import request, error
from bs4 import BeautifulSoup
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Webscraping the website containing the number of wins for each driver (1-25)
url = 'https://pitwall.app/records/driver-wins?page=1'

response = BeautifulSoup(request.urlopen(url), 'html.parser')

TAB = response.find('table')

Wins_df = pd.DataFrame(columns = ['Position','Driver Name','First Win','Last Win','Driver Races','Driver Wins'])

rows = TAB.find_all('tr')

# Inserting the webscraped data into the dataframe
for row in rows[1:]:
    cells = row.find_all('td')
    row_contents = []
    for cell in cells:
        row_contents.append(cell.text.strip())
    length = len(Wins_df)
    Wins_df.loc[length] = row_contents

# Webscraping the website containing the number of wins for the next 25 drivers
url = 'https://pitwall.app/records/driver-wins?page=2'

response = BeautifulSoup(request.urlopen(url), 'html.parser')

TAB = response.find('table')

rows = TAB.find_all('tr')

# Inserting this new data into the same dataframe
for row in rows[1:]:
    cells = row.find_all('td')
    row_contents = []
    for cell in cells:
        row_contents.append(cell.text.strip())
    length = len(Wins_df)
    Wins_df.loc[length] = row_contents

# Dropping unnecessary columns from the dataframe
Wins_df = Wins_df.drop('First Win', axis=1)
Wins_df = Wins_df.drop('Last Win', axis=1)
Wins_df = Wins_df.drop('Position', axis=1)

# Changing the 'Driver Wins' column from a string to an integer
Wins_df['Driver Wins'] = Wins_df['Driver Wins'].astype(int)

# Plotting the bar chart for number of wins
Wins_df.plot.barh(x = 'Driver Name', y = 'Driver Wins', fontsize = 6)

# Inverting the graph so that the drivers with the highest number of wins are listed first
plt.gca().invert_yaxis()

# Adding a title to the graph
plt.title('Top 50 Drivers with the Most F1 Wins')
```

Based on these statistics alone, Lewis Hamilton, having the highest number of wins (103), podiums (197), pole positions (105), and points (4625.5), and equal most world championships with 7, should be considered the most succesful, and therefore greatest, Formula One Driver of all time. It is worth noting that Michael Schumacher, Max Verstappen, Sebastian Vettel, Alain Prost, Ayrton Senna, and Fernando Alonso also feature consistently highly across each of these statistics, placing them in the upper echelon of the most successful Formula One drivers across history. Each of these drivers have at least 2 world championships, 32 wins, 80 podiums, and 22 pole positions.

However, these statistics, whilst important, only provide a surface-level analysis of who the greatest Formula One driver of all time is. This is due to the presence of a myriad of other variables that can impact these statistics, outside of pure driving skill. The main extraneous variable is the strength of the car relative to the rest of the field, and thus how much of each driver's success can be attributed to the speed of the car rather than their driving. Additionally, the fact that there are so many more races on the calendar in recent seasons, with there being 24 races in 2024, compared to just 7 in 1950, means that it is significantly easier for contemporary drivers to achieve more wins, podiums and pole positions than their predecessors. The changing of the points scoring system, with winners being awarded more than triple the amount of points now than they did in 1950, makes this metric largely irrelevant when comparing drivers from different eras. This analysis also fails to justly recognise the driving talent of individuals who had shorter careers, or who passed away while still racing, as it is reasonable to expect that they would have won more races and achieved more pole positions had they raced for longer.

Therefore, before we can conclusively decide who the greatest F1 driver is, we will try to make the comparison fairer by attempting to remove the bias towards drivers who have raced for a longer period of time, and/or completed more races. This will be done by looking at the percentage of races in which each driver achieved wins, podiums and pole positions, and the percentage of years raced in which each driver won world championships.

### Section 2: Identifying the Frequency of Wins, Podiums, Pole Positions and Championships

```python
# Creating a bar chart to demonstrate the frequency of wins for each driver
# For conciseness, only the 10 drivers with the highest win percentage are included in the table 

# Converting "Driver Races" to an integer
Wins_df['Driver Races'] = Wins_df['Driver Races'].astype(int)

# Adding additional column to dataframe containing the percentage of races won by each driver
Wins_df["Win Percentage"] = (Wins_df["Driver Wins"]/Wins_df["Driver Races"]) * 100

# Sorting dataframe so the highest percentage is listed first
Wins_df = Wins_df.sort_values(by=["Win Percentage"])

# Plotting the bar chart for percentage of race wins
Wins_df.plot.barh(x = 'Driver Name', y = 'Win Percentage', fontsize = 6.5)

# Displaying dataframe - limited to top 10
Wins_df = Wins_df.sort_values(by=["Win Percentage"], ascending=False)
display(Wins_df.head(10))
```

```python
# Creating a bar chart to demonstrate the frequency of podiums for each driver
# Only the 10 drivers with the highest podium percentage are included in the table 

# Converting "Driver Races" to an integer
Podiums_df['Driver Races'] = Podiums_df['Driver Races'].astype(int)

# Adding additional column to dataframe containing the percentage of podiums achieved by each driver
Podiums_df["Podium Percentage"] = (Podiums_df["Driver Podiums"]/Podiums_df["Driver Races"]) * 100

# Sorting dataframe so the highest percentage is listed first
Podiums_df = Podiums_df.sort_values(by=["Podium Percentage"])

# Plotting the bar chart for percentage of race wins
Podiums_df.plot.barh(x = 'Driver Name', y = 'Podium Percentage', fontsize = 6.5)

# Displaying dataframe - limited to top 10
Podiums_df = Podiums_df.sort_values(by=["Podium Percentage"], ascending=False)
display(Podiums_df.head(10))
```

```python
# Creating a bar chart to demonstrate the frequency of pole positions for each driver
# Only the 10 drivers with the highest pole position percentage are included in the table 

# Converting "Driver Races" to an integer
Poles_df['Driver Races'] = Poles_df['Driver Races'].astype(int)

# Adding additional column to dataframe containing the percentage of pole positions achieved by each driver
Poles_df["Pole Position Percentage"] = (Poles_df["Driver Pole Positions"]/Poles_df["Driver Races"]) * 100

# Sorting dataframe so the highest percentage is listed first
Poles_df = Poles_df.sort_values(by=["Pole Position Percentage"])

# Plotting the bar chart for percentage of race wins
Poles_df.plot.barh(x = 'Driver Name', y = 'Pole Position Percentage', fontsize = 6.5)

# Displaying dataframe - limited to top 10
Poles_df = Poles_df.sort_values(by=["Pole Position Percentage"], ascending=False)
display(Poles_df.head(10))
```

```python
# Creating a bar chart to demonstrate the frequency of world championships for each driver
# Only the drivers with at least two championships have been included in the table 

# For robustness, re-webscraping the world championships data
url = ' https://pitwall.app/records/driver-championships'
response = BeautifulSoup(request.urlopen(url), 'html.parser')
TAB = response.find('table')
Championships_df = pd.DataFrame(columns = ['Position','Driver Name','Races','Wins','Championships'])
rows = TAB.find_all('tr')

# Inserting the webscraped data into the dataframe
for row in rows[1:]:
    cells = row.find_all('td')
    row_contents = []
    for cell in cells:
        row_contents.append(cell.text.strip())
    length = len(Championships_df)
    Championships_df.loc[length] = row_contents

# Dropping unnecessary columns from the dataframe
Championships_df = Championships_df.drop('Wins', axis=1)
Championships_df = Championships_df.drop('Races', axis=1)
Championships_df = Championships_df.drop('Position', axis=1)

# Changing the 'Driver Pole Positions' column from a string to an integer
Championships_df['Championships'] = Championships_df['Championships'].astype(int)

# Reducing dataframe to only include drivers with at least two world championships
Championships_df = Championships_df.loc[0:16,:]

# Adding additional column to dataframe containing the number of years raced by each driver
Championships_df["Years Raced"] = [18,19,7,16,13,13,16,14,11,10,9,21,11,9,6,11,18]

# Adding additional column to dataframe containing the percentage of podiums achieved by each driver
Championships_df["Championship Percentage"] = (Championships_df["Championships"]/Championships_df["Years Raced"]) * 100

# Sorting dataframe so the highest percentage is listed first
Championships_df = Championships_df.sort_values(by=["Championship Percentage"])

# Plotting the bar chart for percentage of race wins
Championships_df.plot.barh(x = 'Driver Name', y = 'Championship Percentage', fontsize = 8)

# Displaying dataframe
Championships_df = Championships_df.sort_values(by=["Championship Percentage"], ascending=False)
display(Championships_df)
```

This analysis aimed to elimate the effects of differing eras and career lengths in order to make a fairer, more comparable conclusion about which driver is the greatest of all time. Interestingly, Lewis Hamilton, Michael Schumacher, Max Verstappen, Ayrton Senna, and Alain Prost are all again within the top 10 drivers for their frequency of wins, podiums, pole positions and championships, providing extra strength to their case to be considered the greatest F1 driver of all time, as they not only won a significant number of races and world championships, but also did so on a very regular and consistent basis. However, due to their slightly lower win percentages, Fernando Alonso and Sebastian Vettel, despite boasting impressive records, can be considered to be less successful than the other drivers.

Additionally, whilst not recognised in previous analysis due to their relatively short careers, it is now evident that Juan Fangio, Alberto Ascari, Jim Clark, and Jackie Stewart, with each of them winning at least 27% of the races they entered, are also worthy of being discussed amongst the others as exceptional F1 drivers. However, it is worth noting that whilst these drivers have similar, or better, percentages than the other drivers in consideration, as they all raced in 100 races or less, the frequency of their wins carries less weight than Lewis Hamilton or Michael Schumacher, who were able to win approximately 30% of the 300+ races they entered, proving they can race at an extremely high level over a long period of time.

So, in order to get a better idea of which of these drivers is conclusively better than the others, we will attempt to identify the relative strength of the cars driven by these drivers in their championship winning years, thereby removing the variable of the car, which will then generate a clearer picture of which of these hugely succesful driver was the most skillful.

### Section 3: Analysing the Performance of the Drivers Relative to their Teammates

```python
# Bar chart of the gap between drivers champ winner and second place (probably teammate)
# (only discuss best few drivers - averaged if multiple championship winner) - bigger gap is better as well outperformed teammate

# Webscraping the website containing the list of drivers' championships and the margin to second place
url = 'https://en.wikipedia.org/wiki/List_of_Formula_One_World_Drivers%27_Champions'

response = BeautifulSoup(request.urlopen(url), 'html.parser')

TAB = response.find_all('table')[2]

Drivers_Champ_df = pd.DataFrame(columns = ['Season','Driver','Age','Chassis','Engine','Tyres', 'Poles', 'Wins', 'Podiums', 'Fastest Laps', 'Points', 'Percentage of Points', 'Clinched', 'Number of rounds remaining', 'Margin', 'Margin (Percentage)'])

rows = TAB.find_all('tr')

# Inserting the webscraped data into the dataframe
for row in rows[2:7]:
    cells = row.find_all('td')
    individual_row_data = [cell.text.strip() for cell in cells]
    length = len(Drivers_Champ_df)
    Drivers_Champ_df.loc[length] = individual_row_data

for row in rows[8:77]:
    cells = row.find_all('td')
    individual_row_data = [cell.text.strip() for cell in cells]
    length = len(Drivers_Champ_df)
    Drivers_Champ_df.loc[length] = individual_row_data

# Dropping unnecessary columns from the dataframe
Drivers_Champ_df = Drivers_Champ_df.drop('Age', axis=1)
Drivers_Champ_df = Drivers_Champ_df.drop('Engine', axis=1)
Drivers_Champ_df = Drivers_Champ_df.drop('Tyres', axis=1)
Drivers_Champ_df = Drivers_Champ_df.drop('Poles', axis=1)
Drivers_Champ_df = Drivers_Champ_df.drop('Wins', axis=1)
Drivers_Champ_df = Drivers_Champ_df.drop('Podiums', axis=1)
Drivers_Champ_df = Drivers_Champ_df.drop('Fastest Laps', axis=1)
Drivers_Champ_df = Drivers_Champ_df.drop('Points', axis=1)
Drivers_Champ_df = Drivers_Champ_df.drop('Percentage of Points', axis=1)
Drivers_Champ_df = Drivers_Champ_df.drop('Clinched', axis=1)
Drivers_Champ_df = Drivers_Champ_df.drop('Number of rounds remaining', axis=1)
Drivers_Champ_df = Drivers_Champ_df.drop('Margin', axis=1)

# Selecting data from dataframe relating to the 9 drivers in consideration
# Juan Fangio
count = 0
total_margin = 0
for index, row in Drivers_Champ_df.iterrows():
    if row['Driver'] == 'Juan Manuel Fangio[22]':
        count = count + 1
        total_margin = total_margin + float(row['Margin (Percentage)'])
        
Fangio_ave_margin = total_margin/count
Fangio_ave_margin = round(Fangio_ave_margin, 2)

# Lewis Hamilton
count = 0
total_margin = 0
for index, row in Drivers_Champ_df.iterrows():
    if row['Driver'] == 'Lewis Hamilton[52]':
        count = count + 1
        total_margin = total_margin + float(row['Margin (Percentage)'])
        
Hamilton_ave_margin = total_margin/count
Hamilton_ave_margin = round(Hamilton_ave_margin, 2)

# Michael Schumacher
count = 0
total_margin = 0
for index, row in Drivers_Champ_df.iterrows():
    if row['Driver'] == 'Michael Schumacher[45]':
        count = count + 1
        total_margin = total_margin + float(row['Margin (Percentage)'])
        
Schumacher_ave_margin = total_margin/count
Schumacher_ave_margin = round(Schumacher_ave_margin, 2)

# Max Verstappen
count = 0
total_margin = 0
for index, row in Drivers_Champ_df.iterrows():
    if row['Driver'] == 'Max Verstappen[57]' or row['Driver'] == 'Max Verstappen[56]':
        count = count + 1
        total_margin = total_margin + float(row['Margin (Percentage)'])
        
Verstappen_ave_margin = total_margin/count
Verstappen_ave_margin = round(Verstappen_ave_margin, 2)

# Ayrton Senna
count = 0
total_margin = 0
for index, row in Drivers_Champ_df.iterrows():
    if row['Driver'] == 'Ayrton Senna[43]':
        count = count + 1
        total_margin = total_margin + float(row['Margin (Percentage)'])
        
Senna_ave_margin = total_margin/count
Senna_ave_margin = round(Senna_ave_margin, 2)

# Alain Prost
count = 0
total_margin = 0
for index, row in Drivers_Champ_df.iterrows():
    if row['Driver'] == 'Alain Prost[42]':
        count = count + 1
        total_margin = total_margin + float(row['Margin (Percentage)'])
        
Prost_ave_margin = total_margin/count
Prost_ave_margin = round(Prost_ave_margin, 2)

# Alberto Ascari
count = 0
total_margin = 0
for index, row in Drivers_Champ_df.iterrows():
    if row['Driver'] == 'Alberto Ascari[24]':
        count = count + 1
        total_margin = total_margin + float(row['Margin (Percentage)'])
        
Ascari_ave_margin = total_margin/count
Ascari_ave_margin = round(Ascari_ave_margin, 2)

# Jim Clark
count = 0
total_margin = 0
for index, row in Drivers_Champ_df.iterrows():
    if row['Driver'] == 'Jim Clark[29]':
        count = count + 1
        total_margin = total_margin + float(row['Margin (Percentage)'])
        
Clark_ave_margin = total_margin/count
Clark_ave_margin = round(Clark_ave_margin, 2)

# Jackie Stewart
count = 0
total_margin = 0
for index, row in Drivers_Champ_df.iterrows():
    if row['Driver'] == 'Jackie Stewart[32]':
        count = count + 1
        total_margin = total_margin + float(row['Margin (Percentage)'])
        
Stewart_ave_margin = total_margin/count
Stewart_ave_margin = round(Stewart_ave_margin, 2)

# Putting the percentages into a dataframe
data = [['Juan Fangio', Fangio_ave_margin], ['Lewis Hamilton', Hamilton_ave_margin], ['Michael Schumacher', Schumacher_ave_margin], ['Max Verstappen', Verstappen_ave_margin], ['Ayrton Senna', Senna_ave_margin], ['Alain Prost', Prost_ave_margin], ['Alberto Ascari', Ascari_ave_margin], ['Jim Clark', Clark_ave_margin], ['Jackie Stewart', Stewart_ave_margin]]
Drivers_Championship_Margin_df  = pd.DataFrame(data, columns = ['Driver', 'Average Championship Winning Margin (%)'])
Drivers_Championship_Margin_df = Drivers_Championship_Margin_df.sort_values(by=["Average Championship Winning Margin (%)"], ascending=False)
display(Drivers_Championship_Margin_df)

# Creating a bar chart
Drivers_Championship_Margin_df.plot.barh(x = 'Driver', y = 'Average Championship Winning Margin (%)')
plt.gca().invert_yaxis()
plt.title("Average Winning Margin in the Drivers' Championship")
plt.legend(loc = 'lower right', fontsize=8)
```

This graph shows the margin between first and second place in the Drivers' Championship for each of the drivers, averaged out across the years that they won. This therefore hopes to demonstrate the pure driving skill of the drivers, as a higher average winning margin suggests that they won their championships more easily, and so were significantly better than the rest of the field, including their teammates who would have been competing in an identical car. We are thus able to conclude that the driver's success is more attributable to their own talent, rather than the relative strength of the car compared to the rest of the field, when their average championship winning margin is higher.

Based on this analysis, Jackie Stewart, having the highest average championship winning margin of any of the drivers at 36.86%, was clearly signifcantly more talented than the rest of the drivers he was competing against, meaning more of his success is likely due to his skill as a driver, rather than by simply having a faster car than everybody else.
```python
# Analysis of the average margin between first and second place in the constructors' championship, in order to analyse the
# relative strength of the drivers' cars compared to the rest of the field
# Only the drivers in consideration have been included
# Please Note: The dataframe only begins in 1958, as that was when the constructors' championship was introduced. This means
# that Juan Fangio and Alberto Ascari cannot be included in this analysis

# Webscraping the website containing the list of Constructors' championships and the margin to second place
url = 'https://en.wikipedia.org/wiki/List_of_Formula_One_World_Constructors%27_Champions#:~:text=Out%20of%20the%20170%20chassis,the%20title%20on%2016%20occasions.'

response = BeautifulSoup(request.urlopen(url), 'html.parser')

TAB = response.find_all('table')[2]

Constructors_Champ_df = pd.DataFrame(columns = ['Season','Chassis','Engine','Tyre','Drivers','Poles', 'Wins', 'Podiums', 'Fastest Laps', 'Points', 'Percentage of Points', 'Clinched', 'Margin (Points)', 'Margin (Percentage)'])

rows = TAB.find_all('tr')

# Inserting the webscraped data into the dataframe
for row in rows[2:68]:
    cells = row.find_all('td')
    individual_row_data = [cell.text.strip() for cell in cells]
    length = len(Constructors_Champ_df)
    Constructors_Champ_df.loc[length] = individual_row_data

# Dropping unnecessary columns from the dataframe
Constructors_Champ_df = Constructors_Champ_df.drop('Engine', axis=1)
Constructors_Champ_df = Constructors_Champ_df.drop('Tyre', axis=1)
Constructors_Champ_df = Constructors_Champ_df.drop('Poles', axis=1)
Constructors_Champ_df = Constructors_Champ_df.drop('Wins', axis=1)
Constructors_Champ_df = Constructors_Champ_df.drop('Podiums', axis=1)
Constructors_Champ_df = Constructors_Champ_df.drop('Fastest Laps', axis=1)
Constructors_Champ_df = Constructors_Champ_df.drop('Points', axis=1)
Constructors_Champ_df = Constructors_Champ_df.drop('Percentage of Points', axis=1)
Constructors_Champ_df = Constructors_Champ_df.drop('Clinched', axis=1)
Constructors_Champ_df = Constructors_Champ_df.drop('Margin (Points)', axis=1)

# Cleaning data by removing extra unnecessary symbols
Constructors_Champ_df = Constructors_Champ_df.replace("†","", regex=True)


# Selecting data from dataframe relating to the 9 drivers in consideration
# Jackie Stewart
count = 0
total_margin = 0
for index, row in Drivers_Champ_df.iterrows():
    if row['Driver'] == 'Jackie Stewart[32]':
        winning_season = row['Season']
        count = count + 1
        for index2, row2 in Constructors_Champ_df.iterrows():
            if row2['Season'] == winning_season:
                total_margin = total_margin + float(row2['Margin (Percentage)'])
        
Stewart_ave_margin = total_margin/count
Stewart_ave_margin = round(Stewart_ave_margin, 2)

# Lewis Hamilton
count = 0
total_margin = 0
for index, row in Drivers_Champ_df.iterrows():
    if row['Driver'] == 'Lewis Hamilton[52]':
        winning_season = row['Season']
        count = count + 1
        for index2, row2 in Constructors_Champ_df.iterrows():
            if row2['Season'] == winning_season:
                total_margin = total_margin + float(row2['Margin (Percentage)'])
        
Hamilton_ave_margin = total_margin/count
Hamilton_ave_margin = round(Hamilton_ave_margin, 2)

# Michael Schumacher
count = 0
total_margin = 0
for index, row in Drivers_Champ_df.iterrows():
    if row['Driver'] == 'Michael Schumacher[45]':
        winning_season = row['Season']
        count = count + 1
        for index2, row2 in Constructors_Champ_df.iterrows():
            if row2['Season'] == winning_season:
                total_margin = total_margin + float(row2['Margin (Percentage)'])
        
Schumacher_ave_margin = total_margin/count
Schumacher_ave_margin = round(Schumacher_ave_margin, 2)

# Max Verstappen
count = 0
total_margin = 0
for index, row in Drivers_Champ_df.iterrows():
    if row['Driver'] == 'Max Verstappen[57]' or row['Driver'] == 'Max Verstappen[56]':
        winning_season = row['Season']
        count = count + 1
        for index2, row2 in Constructors_Champ_df.iterrows():
            if row2['Season'] == winning_season:
                total_margin = total_margin + float(row2['Margin (Percentage)'])
        
Verstappen_ave_margin = total_margin/count
Verstappen_ave_margin = round(Verstappen_ave_margin, 2)

# Ayrton Senna
count = 0
total_margin = 0
for index, row in Drivers_Champ_df.iterrows():
    if row['Driver'] == 'Ayrton Senna[43]':
        winning_season = row['Season']
        count = count + 1
        for index2, row2 in Constructors_Champ_df.iterrows():
            if row2['Season'] == winning_season:
                total_margin = total_margin + float(row2['Margin (Percentage)'])
        
Senna_ave_margin = total_margin/count
Senna_ave_margin = round(Senna_ave_margin, 2)

# Alain Prost
count = 0
total_margin = 0
for index, row in Drivers_Champ_df.iterrows():
    if row['Driver'] == 'Alain Prost[42]':
        winning_season = row['Season']
        count = count + 1
        for index2, row2 in Constructors_Champ_df.iterrows():
            if row2['Season'] == winning_season:
                total_margin = total_margin + float(row2['Margin (Percentage)'])
        
Prost_ave_margin = total_margin/count
Prost_ave_margin = round(Prost_ave_margin, 2)

# Jim Clark
count = 0
total_margin = 0
for index, row in Drivers_Champ_df.iterrows():
    if row['Driver'] == 'Jim Clark[29]':
        winning_season = row['Season']
        count = count + 1
        for index2, row2 in Constructors_Champ_df.iterrows():
            if row2['Season'] == winning_season:
                total_margin = total_margin + float(row2['Margin (Percentage)'])
        
Clark_ave_margin = total_margin/count
Clark_ave_margin = round(Clark_ave_margin, 2)

# Putting the percentages into a dataframe
data = [['Lewis Hamilton', Hamilton_ave_margin], ['Michael Schumacher', Schumacher_ave_margin], ['Max Verstappen', Verstappen_ave_margin], ['Ayrton Senna', Senna_ave_margin], ['Alain Prost', Prost_ave_margin], ['Jim Clark', Clark_ave_margin], ['Jackie Stewart', Stewart_ave_margin]]
Constructors_Championship_Margin_df  = pd.DataFrame(data, columns = ['Driver', 'Average Championship Winning Margin (%)'])
Constructors_Championship_Margin_df = Constructors_Championship_Margin_df.sort_values(by=["Average Championship Winning Margin (%)"], ascending=False)
display(Constructors_Championship_Margin_df)

# Creating a bar chart
Constructors_Championship_Margin_df.plot.barh(x = 'Driver', y = 'Average Championship Winning Margin (%)')
plt.gca().invert_yaxis()
plt.title("Average Winning Margin in the Constructors' Championship")
plt.legend(loc = 'lower right', fontsize=8)
```
